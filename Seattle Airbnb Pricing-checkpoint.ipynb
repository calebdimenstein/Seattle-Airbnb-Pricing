# Data Import

```python
'''
Airbnb dataframe lists all the details of each house that's availble to rent

Time_Available is used to see how many days in 2016 the property was availble for. This is used to 
measure its disirability
'''

Airbnb = pd.read_excel('C:\\Users\\caldimen1\\Desktop\\UChicago\\Statistical Analysis\\Tableau Full Project.xlsx',
                      sheet_name = 'Listings')

Time_Available = pd.read_excel('C:\\Users\\caldimen1\\Desktop\\UChicago\\Statistical Analysis\\Tableau Full Project.xlsx',
                      sheet_name = 'Calendar')
```

```python
Airbnb = Airbnb.drop(columns = {'space','listing_url', 'scrape_id', 'last_scraped','host_id', 'host_url', 'experiences_offered', 'notes',
                                'host_acceptance_rate', 'host_thumbnail_url', 'transit', 'medium_url', 'host_url', 'host_about',
                                'host_listings_count','host_picture_url','host_verifications', 'street', 'smart_location', 'country_code', 
                                'country', 'jurisdiction_names','license', 'neighbourhood_group_cleansed', 'city', 'state', 'zipcode',
                                'market', 'license'})
Airbnb = Airbnb.rename(columns = {'id' : 'listing_id'})

Airbnb = Airbnb[Airbnb['listing_id'] != 75131198]
```

```python
'''
Merging the calendar dataframe of how long each listing was available for with the the property details dataframe
'''
grouped = Time_Available.groupby('listing_id')

def calculate_counts(group):
    true = (group['available'] == 't').sum()
    false = (group['available'] == 'f').sum()
    return pd.Series({'Days_Available' : true, 'Days_Booked' : false})
grouped_listing = grouped.apply(calculate_counts)

Airbnb = Airbnb.merge(grouped_listing, on = 'listing_id', how = 'inner')
```

## Checking for Heterosketasticity

```python
'''
Checking for heterosketasticity for accommodates to price
'''

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].scatter(x = ModelDf['accommodates'], y = ModelDf['price'], label = 'Scatter Plot', s = 10)
slope, intercept, r_value, p_value, std_err = linregress(ModelDf['accommodates'], ModelDf['price'])
axs[0].plot(ModelDf['accommodates'], intercept + slope * ModelDf['accommodates'], color='red', label='Linear Regression')
axs[0].set_xlabel('accommodates')
axs[0].set_ylabel('price')
axs[0].set_title('Scatter Plot Accommodates vs Price')
axs[0].legend()

predicted_values = intercept + slope * ModelDf['accommodates']
residuals = ModelDf['price'] - predicted_values

# Plot residuals
axs[1].scatter(ModelDf['accommodates'], residuals, label='Residuals', s = 10)
axs[1].set_xlabel('accommodates')
axs[1].set_ylabel('residuals')
axs[1].set_title('Scatter Plot Accommodate vs residuals')
axs[1].legend()
# Add a horizontal line at y=0 for reference
axs[1].axhline(y=0, color='red', linestyle='--', label='Zero Residuals')
plt.tight_layout()

plt.savefig('accommodates to price.jpg', format='jpeg', dpi=300)  # Adjust the file name and DPI as needed
```

## F Tests

```python
no_ammen = 'price ~ accommodates + bathrooms + accommodates:bathrooms+ beds_to_accommodates + review_scores_rating + cleaning_fee_pct + beds_to_accommodates + Days_Booked + Apartment + House + Townhouse + Condominium'
model_no_amen = smf.ols(formula = no_ammen, data = ModelDf).fit()
print(model_no_amen.summary())
model_no_amen.fvalue
model_no_amen.f_pvalue

no_house = 'price ~ accommodates + bathrooms + accommodates:bathrooms+ beds_to_accommodates + review_scores_rating + cleaning_fee_pct + beds_to_accommodates + Days_Booked + Family_kid_friendly +  Hot_Tub + Safety_Card + washer_dryer + Indoor_Fire_Place  + Pets_Allowed'
model_no_house = smf.ols(formula = no_house, data = ModelDf).fit()
print(model_no_house.summary())
model_no_amen.fvalue
model_no_amen.f_pvalue

no_days_booked = 'price ~ accommodates + bathrooms + accommodates:bathrooms+ beds_to_accommodates + review_scores_rating + cleaning_fee_pct + beds_to_accommodates + Family_kid_friendly +  Hot_Tub + Safety_Card + washer_dryer + Indoor_Fire_Place  + Pets_Allowed + Apartment + House + Townhouse + Condominium'
model_no_booked = smf.ols(formula = no_days_booked, data = ModelDf).fit()
print(model_no_booked.summary())
model_no_booked.fvalue
model_no_booked.f_pvalue
```

```python

model = smf.ols(formula = fullmodel, data = ModelDf).fit()

#Performing an F test to see if the model with no ameneties is a better fit
anova_table_ammen = anova_lm(model_no_amen, model)
f_statistic_ammen = anova_table_ammen['F'][1]  # F-statistic for the model
p_value_ammen = anova_table_ammen['Pr(>F)'][1]


anova_table_house = anova_lm(model_no_house, model)
f_statistic_house = anova_table_house['F'][1]  # F-statistic for the model
p_value_house = anova_table_house['Pr(>F)'][1]



anova_table_book = anova_lm(model_no_booked, model)
f_statistic_booked = anova_table_book['F'][1]  # F-statistic for the model
p_value_booked = anova_table_book['Pr(>F)'][1]

results_data = {
    'Model Exclusion': ['Ammeneties', 'Property Types', 'Days Booked'],
    'F-Statistic': [f_statistic_ammen, f_statistic_house, f_statistic_booked],
    'P-Value': [p_value_ammen, p_value_house, p_value_booked],
}
#print(results_data)
F_test_df = pd.DataFrame(results_data)
F_test_df = F_test_df.set_index('Model Exclusion')
print(F_test_df.head())
```
